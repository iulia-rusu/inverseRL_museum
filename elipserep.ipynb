```python
def select_files_by_index(museum_df, index_list):
    """
    Return subset of museum_df corresponding to selected indexes.

    Parameters
    ----------
    museum_df : pd.DataFrame
        DataFrame with columns ['file', 'time', 'x', 'y', 'state', 'visibility']. and others
    index_list : list of int
        List of row indexes to select.

    Returns
    -------
    pd.DataFrame
        Subset of museum_df containing only rows at the given indexes.
    """
    # Ensure indexes are valid and unique
    index_list = sorted(set(index_list))
    selected_df = museum_df.loc[index_list].reset_index(drop=True)
    return selected_df
```

```python
selected_rows = select_files_by_index(museum_df, [2, 3, 12])
```

```python
# choose elipse parameters
xs = np.asarray(selected_rows['x'][0] + selected_rows['x'][1] + selected_rows['x'][2])
ys = np.asarray(selected_rows['y'][0] + selected_rows['y'][1] + selected_rows['y'][2])

# 1. Choose center as mean of points
h = xs.mean()
k = ys.mean()

# 2. Choose ellipse radii as half the range in each direction
a = (xs.max() - xs.min()) / 2.0  # x-radius
b = (ys.max() - ys.min()) / 2.0  # y-radius
```

```python
def point_to_ellipse_angle(x, y, h, k, a, b):
    u = (x - h) / a
    v = (y - k) / b
    theta = np.arctan2(v, u)
    return np.mod(theta, 2 * np.pi)

# create a new column first (optional)
selected_rows['thetas'] = None

for row in range(len(selected_rows)):
    xs = np.asarray(selected_rows.iloc[row]['x'])
    ys = np.asarray(selected_rows.iloc[row]['y'])

    thetas = point_to_ellipse_angle(xs, ys, h, k, a, b)

    # assign *into this specific row*
    selected_rows.at[row, 'thetas'] = thetas
```

```python
def plot_traj_with_ellipse(selected_rows, row, h, k, a, b):
    # grab this trajectory
    xs = np.asarray(selected_rows.iloc[row]['x'])
    ys = np.asarray(selected_rows.iloc[row]['y'])
    thetas = np.asarray(selected_rows.iloc[row]['thetas'])

    # make ellipse outline
    grid = np.linspace(0, 2*np.pi, 400)
    ell_x = h + a * np.cos(grid)
    ell_y = k + b * np.sin(grid)

    fig, ax = plt.subplots(figsize=(5, 5))

    # trajectory points colored by theta
    sc = ax.scatter(xs, ys, c=thetas, cmap='hsv', s=15)
    ax.plot(ell_x, ell_y, 'k--', lw=1.5)

    ax.set_aspect('equal', 'box')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(f'Trajectory {row} on ellipse (color = )')

    cbar = fig.colorbar(sc, ax=ax)
    cbar.set_label(' (radians)')

    plt.tight_layout()
    plt.show()
```

```python
plot_traj_with_ellipse(selected_rows, row=0, h=h, k=k, a=a, b=b)
```

```python
def plot_theta_vs_time(selected_rows, row):
    t = np.asarray(selected_rows.iloc[row]['time'])
    thetas = np.asarray(selected_rows.iloc[row]['thetas'])

    fig, ax = plt.subplots(figsize=(6, 3))
    ax.plot(t, thetas, marker='.', linestyle='-')
    ax.set_xlabel('time (s)')
    ax.set_ylabel(' (radians)')
    ax.set_title(f'(t) for trajectory {row}')
    plt.tight_layout()
    plt.show()
```

```python
plot_theta_vs_time(selected_rows, 0)
```
