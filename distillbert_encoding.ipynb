```python
# looking into the model architecture
# Look at the first transformer layer
layer0 = model.transformer.layer[0]

# Look at its attention module
attn = layer0.attention

print(attn)
#Q, K, V are concatenated across heads 
```

```python
B, T = 2, 10
vocab = config.vocab_size  # 30522

input_ids = torch.randint(0, vocab, (B, T))
attention_mask = torch.ones(B, T, dtype=torch.long)

out = model(input_ids=input_ids, attention_mask=attention_mask)
print(out.last_hidden_state.shape)
```
